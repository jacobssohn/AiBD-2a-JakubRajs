```python
import pandas as pd
original_data = pd.read_csv('../Original data/earthquake_data.csv')
```

Original data is presented below.

```python
original_data.head()
```

Original names of the columns have been shortened as follows:

```python
new_names = {
    original_data.columns[0]: 'Worried about earthquakes',
    original_data.columns[1]: 'Worried about the Big One',
    original_data.columns[2]: 'Will the Big One occur in your life?',
    original_data.columns[3]: 'Experienced an earthquake?',
    original_data.columns[4]: 'Taken precautions?',
    original_data.columns[5]: 'Familiar with SA Fault line?',
    original_data.columns[6]: 'Familiar with the Yellowstone Supervolcano?',
    original_data.columns[8]: 'Gender',
    original_data.columns[9]: 'Household income'}


new_data = original_data.rename(columns=new_names)
new_data.dropna(subset = ['Age', 'Gender'], inplace=True)
display(new_data)
print(f'The number of deleted records with NaN values: {original_data.shape[0] - new_data.shape[0]}')
```

In the next table only 3 columns will be displayed: age, gender and the answer to the question whether the Big One will occur in the person's lifetime.

```python
small_data = new_data[[new_data.columns[7], new_data.columns[8], new_data.columns[2]]]
display(small_data)
```

```python
new_data.to_csv('../Analysis data/new_data.csv')
small_data.to_csv('../Analysis data/small_data.csv')
```
